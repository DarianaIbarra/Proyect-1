```python
# Second step: We're going to read our Resources.csv and test to see if we can see it.
SNP_file = pd.read_csv('Resources/Input_Files/GWAS_Parkinson.csv')    #You should always open the Data_Extraction_Code 
                                                       #folder to have a relative path similar to this one
SNP_file.head(30)
```

```python
# Third step: Now let's create a second dataframe (to keep from editing the original)
SNP_editable = pd.DataFrame(SNP_file)
SNP_editable.head()
```

```python
# Now we have our own database set up as a dataframe and filled it with our columns. 
output_path = 'Resources/OutputFromExtractionCode'
print(SNP_editable.columns.tolist())
```

```python
# Now let's drop all the data we collected and just stay with the two things that matter most to us: 
# the SNP ID and the Gene:
SNP_ID_And_Gene_From_Data = SNP_editable[['MAPPED_GENE','SNP ID','DISEASE/TRAIT','CHR_ID','P-VALUE','INITIAL SAMPLE SIZE']]
SNP_ID_And_Gene_From_Data.head(10)
```

```python
### Now we can't use anything that does not have a standard SNP ID (i.e. rs#) So we need to drop them.
identifier = 'rs'

snps_to_keep = [column for column in SNP_ID_And_Gene_From_Data if identifier in column]
SNP_ID_And_Gene_From_Data = SNP_ID_And_Gene_From_Data[SNP_ID_And_Gene_From_Data['SNP ID'].str.contains(identifier, na=False)]
SNP_ID_And_Gene_From_Data.head(20)
```

```python
### We'll be using this for our RENTREZ module in R'. Let's output it twice (one for keeps and one we can modify safely)
SNP_ID_And_Gene_From_Data.to_csv(os.path.join(output_path, 'Original_Gene_And_dbSNP_ID.csv'), index=False)
SNP_ID_And_Gene_From_Data.to_csv(os.path.join(output_path, 'Editable_Gene_And_dbSNP_ID.csv'), index=False)
```

```python
### We can already do something pretty cool, which is determine which gene had more pathogenic SNPs.
### We can sort it:
SNP_Sorted_By_Gene =  SNP_ID_And_Gene_From_Data.sort_values(by='MAPPED_GENE')
SNP_Sorted_By_Gene.head(10)
```

```python
### And get the unique count for each gene:
Unique_Genes = SNP_Sorted_By_Gene['MAPPED_GENE'].value_counts(normalize=False, dropna=True).reset_index()
Unique_Genes.columns = ['MAPPED_GENE', 'SNP ID']
Unique_Genes.head(20)
```

```python
### And we should group them by gene so we can see which SNPs are actually related:
grouped_SNPs = SNP_Sorted_By_Gene.groupby('MAPPED_GENE')['SNP ID'].apply(list).reset_index()
grouped_SNPs.head(20)

# groupby('category') groups the data by the 'category' column.
# apply(list) applies the list function to each group, resulting in a list of strings for each unique category.
# reset_index() is used to turn the result back into a DataFrame.
```

```python
### Now let's output them as CSV files:
SNP_Sorted_By_Gene.to_csv(os.path.join(output_path, 'SNP_Sorted_By_Gene.csv'), index=False)
Unique_Genes.to_csv(os.path.join(output_path, 'Unique_Genes.csv'), index=False)
grouped_SNPs.to_csv(os.path.join(output_path, 'grouped_SNPs.csv'), index=False)
```

```python
### Now let's get the population samples
Split_Pop = SNP_Sorted_By_Gene.copy()
Split_Pop.head(50)

# Split the column into multiple columns
split_data = Split_Pop['INITIAL SAMPLE SIZE'].str.split(' ', expand=True)

# Ensure exactly 4 columns by reindexing
split_data = split_data.reindex(columns=range(5))  # Create 4 columns, adding None/NaN if necessary

# Assign the split columns back to the DataFrame
Split_Pop[['Samples', 'Ethnicity/Origin', 'Split3', 'Split4', 'Split5']] = split_data

# Optional: Check the result
Split_Pop
```

```python
#Now let's drop the other split columns and identify all of our values to see if we need to clean up
Population_Data = Split_Pop[['MAPPED_GENE', 'SNP ID', 'Samples', 'Ethnicity/Origin']]
Population_Data_List = (Population_Data['Ethnicity/Origin'].value_counts())
Population_Data_List
```

```python
###Let's make a copy
Population_Data_Clean = Population_Data.copy()

### Let's change the following values:
#[1] 'Cases,', 'individuals', 'cases' -->> 'Not Reported'
word_finder_unknown = ['cases,', 'individuals', 'cases']
#[2] 'East' --->>> 'East Asian'
word_finder_east = ['east']
#[3] 'Han' --->>> 'Chinese'
word_finder_han = ['han']
#[4] 'South' --->>> 'South Asian'
word_finder_south = ['south']

## Now let's make a function to change ethnicity
def change_ethnicity(value):
    if any(word in value.lower() for word in word_finder_unknown):
        return 'Not Reported'
    elif any(word in value.lower() for word in word_finder_east):
        return 'East Asian'
    elif any(word in value.lower() for word in word_finder_han):
        return 'Chinese'
    elif any(word in value.lower() for word in word_finder_south):
        return 'South Asian'
    else:
        return value

## Let's apply it to our function
Population_Data_Clean['Ethnicity/Origin'] = Population_Data_Clean['Ethnicity/Origin'].apply(change_ethnicity)

## Let's make it a series and check if there's any other population
Population_Data_List_Clean = (Population_Data_Clean['Ethnicity/Origin'].value_counts().reset_index)
Population_Data_List_Clean
```

```python
### Let's make it a dataframe
Population_Data_Clean_df = Population_Data_Clean['Ethnicity/Origin'].value_counts(normalize=False, dropna=True).reset_index()
Population_Data_Clean_df
```

```python
### Let's output it as a csv file
Population_Data_Clean_df.to_csv(os.path.join(output_path, 'SNP_count_per_population.csv'), index=False)
```

```python
### let's make a bar chart
plt.close('all')

bar_categories = Population_Data_Clean_df['Ethnicity/Origin']
bar_values = Population_Data_Clean_df['count']
len_bar_categories = len(bar_categories)
width = 5 / len_bar_categories

plt.barh(bar_categories, bar_values,  color= 'red')
plt.ylabel('Ethnicity')
plt.xticks(rotation=90, ha='right')
plt.xlabel('# of SNP reported')
plt.title('Single nucleotide polymorphisms reported per population')

### Let's save it
plt.tight_layout()
output_path_png = 'Resources/OutputFromExtractionCode/SNP_counts_by__ethnicity.png'
plt.savefig(output_path_png, dpi=300,bbox_inches='tight')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.close('all')

# Sort the DataFrame by the count column in descending order
Population_Data_Clean_df = Population_Data_Clean_df.sort_values(by='count', ascending=False)

# Set categories and total values
bar_categories = Population_Data_Clean_df['Ethnicity/Origin']
counts = Population_Data_Clean_df['count']
total_count = counts.sum()

# Waffle chart settings
grid_size = 28
total_tiles = grid_size * grid_size
category_colors = [
    '#1f77b4', '#aec7e8', '#ffbb78', '#98df8a', '#d62728',
    '#9467bd', '#8c564b', '#c49c94', '#f7b6d2', '#c7c7c7',
    '#bcbd22', '#17becf', '#9edae5'
]

# Calculate tiles per category with rounding correction
fractional_tiles = {category: value / total_count * total_tiles for category, value in zip(bar_categories, counts)}
rounded_tiles = {category: int(np.round(tiles)) for category, tiles in fractional_tiles.items()}

# Ensure the total number of tiles matches `total_tiles`
rounded_tiles_adjusted = rounded_tiles.copy()
while sum(rounded_tiles_adjusted.values()) != total_tiles:
    diff = total_tiles - sum(rounded_tiles_adjusted.values())
    largest_category = max(rounded_tiles_adjusted, key=lambda k: fractional_tiles[k] - rounded_tiles_adjusted[k])
    rounded_tiles_adjusted[largest_category] += diff

tiles_per_ethn = rounded_tiles_adjusted

# Create the waffle grid
waffle = np.zeros(total_tiles, dtype=int)

# Fill the waffle grid
category_index = 1
tile_index = 0
for category, tile_count in tiles_per_ethn.items():
    waffle[tile_index:tile_index + tile_count] = category_index
    tile_index += tile_count
    category_index += 1

# Reshape the waffle grid into the specified size
waffle = waffle.reshape((grid_size, grid_size))

# Plot the waffle chart
fig, ax = plt.subplots(figsize=(8, 6))

# Use the new colormap access
cmap = plt.colormaps['tab20']
im = ax.imshow(waffle, cmap=cmap)

# Draw white gridlines explicitly
for x in range(grid_size + 1):
    ax.axhline(x - 0.5, color='white', linewidth=2, zorder=10)  # Horizontal lines
    ax.axvline(x - 0.5, color='white', linewidth=2, zorder=10)  # Vertical lines

# Remove major ticks and labels
ax.tick_params(which='both', bottom=False, left=False, labelbottom=False, labelleft=False)

# Add category legend
handles = [plt.Rectangle((0, 0), 1, 1, color=category_colors[i]) for i in range(len(bar_categories))]
labels = bar_categories
ax.legend(
    handles, labels,
    loc='center left',                # Place legend to the right
    bbox_to_anchor=(1.05, 0.5),       # Adjust positioning
    ncol=1,                           # Single column legend
    title='Ethnicity/Origin'          # Add a title to the legend
)

# Remove axes
ax.axis('off')

# Add a title
#plt.title('Single nucleotide polymorphisms reported per population', loc='lower center', fontsize=14)
fig.suptitle(
    'Single nucleotide polymorphisms reported per population',
    fontsize=14, x=0.45,
    y=0.00  # Adjust the vertical position
)

# Adjust layout for better fit
plt.tight_layout()
output_path_png = 'Resources/OutputFromExtractionCode/SNP_counts_by__ethnicity_waffle.png'
plt.savefig(output_path_png, dpi=300,bbox_inches='tight')
plt.show()

```
